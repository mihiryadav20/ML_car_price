```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
car_ds = pd.read_csv('archive/car data.csv')
```

```python
car_ds.head()
```

```python
car_ds.shape
```

```python
car_ds.info()
```

```python
car_ds.isnull().sum()
```

```python
print(car_ds.Fuel_Type.value_counts())
```

```python
print(car_ds.Seller_Type .value_counts())
```

```python
print(car_ds.Transmission.value_counts())
```

```python
car_ds.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
```

```python
car_ds.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
```

```python
car_ds.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_ds.head()
```

```python
X = car_ds.drop(['Car_Name','Selling_Price'],axis=1)
```

```python
print(X)
```

```python
Y = car_ds["Selling_Price"]
```

```python
print(Y)
```

# splitting the train and test data set

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
print(X_train)
```

# model using linear regression


```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train,Y_train)
```

```python
training_data_prediction = lr_model.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
test_data_prediction = lr_model.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price").set_color('white')
plt.ylabel("Predicted Price").set_color('white')
plt.title(" Actual Prices vs Predicted Prices").set_color('white')
plt.show() 
```

# model using lasso regression

```python
#Lasso regression
lr2=Lasso()
```

```python
lr2.fit(X_train,Y_train)
```

```python
training_data_prediction = lr2.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
test_data_prediction = lr2.predict(X_test)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
